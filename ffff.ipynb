```python
import numpy as np
kr, en, math, sci = map(int,input(" : ").split())

def score(x):
    a=max(x)
    b=min(x)
    c=np.mean(x)
    return a,b,c

a, b,c=score([kr,en,math,sci])
a1,b1,c1=score([en,math])
print(a,b,c)
print(a1,b1,c1)
```

```python
def hl(cnt):
    if cnt==0:
        return
    print("hello",cnt)

    cnt-=1
    hl(cnt)
hl(5)
```

```python
def fac(n):
    if n == 1:
        return 1
    return n*fac(n-1)
fac(4)
```
